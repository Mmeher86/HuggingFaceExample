```python
from transformers import pipeline
sentiment_pipeline = pipeline("sentiment-analysis")
data = ["I love you", "I hate you","Neither i love nor i hate you"]
sentiment_pipeline(data)
```

```python
from transformers import pipeline
sentiment_pipeline = pipeline(model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
data = ["I love you", "I hate you","Neither i love nor i hate you"]
sentiment_pipeline(data)

```

```python
sentiment_analysis_tabular=pipeline(model="tabularisai/multilingual-sentiment-analysis")
sentiment_analysis_tabular(data)
```

```python
sentiment_analysis_finite=pipeline(model="finiteautomata/bertweet-base-sentiment-analysis")
sentiment_analysis_finite(data)
```

```python
sentiment_analysis_Flavio=pipeline(model="cardiffnlp/twitter-roberta-base-sentiment-latest")
sentiment_analysis_Flavio(data)
```
