```python
from transformers import pipeline

question_answerer = pipeline("question-answering", model='distilbert-base-cased-distilled-squad')


context = r"""
India is the biggest democratic country in the world.The capital of India is New Delhi.
It celebarte multiple regional and national festival.National festival of india are Republic day ,Independence day,Gandhi jayanti.
India have 26 states and 6 union teritory.
"""

threshold = 0.5

question=["Biggest democratic country in the world","Which is the capital of India","India have how many union teritory","India have how many states","Is Republic day regional festival"]
for question1 in question:
  result = nlp(question=question1,context=context)
  if result['score'] < threshold:
    print(f"Question: {question1}: Answer: 'Don't know' Score:{result['score']} '{result['answer']}' ")
  else:
    print(f"Question: {question1}: Answer: '{result['answer']}', score: {round(result['score'], 4)}, start: {result['start']}, end: {result['end']}")
```

```python
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering
import torch

tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-cased-distilled-squad')
model = DistilBertForQuestionAnswering.from_pretrained('distilbert-base-cased-distilled-squad')
```

```python

question, text = "Who was Jim Henson?", "Jim Henson was a nice puppet"
input_ids = tokenizer(question, text, return_tensors='pt')

with torch.no_grad():
  output = model(**input_ids)
print(output)



```
